```
# Reading logs files

logs = sc.textFile("/FileStore/logs/")
```

```
# Using map-reduce, count the number of requests from each user

usercounts = logs.map(lambda line: line.split()).map(lambda words: (words[2],1)).reduceByKey(lambda count1,count2: count1 + count2)
usercounts.take(5)
```

```
#Use countByKey to determine how many users visited the site for each frequency. That is, how many users visited
#once, twice, three times and so on.

freq = usercounts.map(lambda x: (x[1],x[0]))
freq_count= freq.countByKey()
print('RDD Count by key:', freq.countByKey())
```

```
#Create an RDD where the user id is the key, and the value is the list of all the IP addresses that user has connected
#from. (IP address is the first field in each request line.)

ipAdd_rdd = logs.map(lambda x : x.split()).map(lambda x: (x[2], x[0])).groupByKey()

for (userid,ips) in ipAdd_rdd.take(10): 
  print (userid, ":" )
  for ip in ips:
    print ("\t",ip )
```

```
#Create an RDD based on the accounts data consisting of key/value-array pairs:

accounts = sc.textFile("/FileStore/accounts").map(lambda s: s.split(',')).map(lambda account: (account[0],account))
accounts.collect()
```

```
#Join the Pair RDD with the set of user-id/hit-count pairs calculated in the previous step.
accounthits = accounts.join(usercounts)

# Display the user ID, hit count, and first name (3rd value) and last name (4th value) for the first 5 elements
for (userid,(values,count)) in accounthits.take(10) : 
  print(userid, count, values[3],values[4])
```
